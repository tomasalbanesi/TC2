```python
# Importacion de librerias a utilizar en todo el notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import sympy as sp
from IPython.display import IFrame
from pytc2.sintesis_dipolo import foster, cauer_LC
from pytc2.dibujar import dibujar_foster_serie, dibujar_foster_derivacion, dibujar_cauer_LC
from pytc2.general import print_latex

# Inicializacion de gráficos
plt.figure(1)
plt.close(1)

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})
```

### 1.a. Resolución Punto 1) a) <a class="anchor" id="1.a"></a>

#### Resolución analítica

#### Resolución simbólica

```python
# Resolución simbólica
s = sp.symbols('s', complex=True)

# Sea la siguiente función de excitación
Z = (s**2+3)*(s**2+1)/(s*(s**2+2))

# Z según Foster derivación

# Implementaremos Imm mediante Foster
k0, koo, ki = foster(1/Z)

dibujar_foster_derivacion(k0, koo, ki, y_exc = 1/Z)
```

### 1.b. Resolución Punto 1) b) <a class="anchor" id="1.b"></a>

#### Resolución analítica Cauer 1

```python
# Visualizacion de PDF
IFrame("./docs/DOC_20231009_0002.pdf", width=1000, height=1000)
```

#### Resolución simbólica Cauer 1

```python
# Resolución simbólica
s = sp.symbols('s', complex=True)

# Sea la siguiente función de excitación
Z = (s**2+3)*(s**2+1)/(s*(s**2+2))

# Z según Cauer 1 (removiendo en oo) 

koo, imm_cauer_oo, rem = cauer_LC(Z, remover_en_inf=True)

print_latex( r'$' + sp.latex(Z) + r'=' + sp.latex(imm_cauer_oo) + r'$' )

# Tratamos a nuestra función inmitancia como una Z
dibujar_cauer_LC(koo, z_exc = imm_cauer_oo)
```

#### Resolución simbólica Cauer 2

```python
# Resolución simbólica
s = sp.symbols('s', complex=True)

# Sea la siguiente función de excitación
Z = (s**2+3)*(s**2+1)/(s*(s**2+2))

# Z según Cauer 2 (removiendo en 0) 

k0, imm_cauer_0, rem = cauer_LC(Z, remover_en_inf=False)

print_latex( r'$' + sp.latex(Z) + r'=' + sp.latex(imm_cauer_0) + r'$' )

# Tratamos a nuestra función inmitancia como una Z
dibujar_cauer_LC(k0, z_exc = imm_cauer_0)
```
