#### Punto 1 - Filtro PASA BAJOS

A continuacion se muestra la ecuacion generica de un filtro pasabajos:

\begin{equation}
H(s) = \frac{\omega_0^2}{s^2+s\frac{\omega_0}{Q}+\omega_0^2}
\end{equation}

siendo $\omega_0$ la frecuencia de corte del filtro y $Q$ el factor de selectividad.

Ahora, para graficarlo le daremos los siguientes valores a la frecuencia y al factor de selectividad:

| $Q$ | $\omega_0$ | Tipo |
| :--- | :--- | :--- |
| $0.1$ | $10$ | Sobreamortiguado |
| $0.5$ | $10$ | Amortiguado crítico |
| $2$ | $10$| Subamortiguado |

A continuacion se procedera a realizar los graficos correspondientes mediante Python:

```python
# Importacion de librerias a utilizar
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
from scipy.signal import TransferFunction
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Esta parte de código la agregamos SOLO en los notebooks para fijar el estilo de los gráficos.
#
#%%  Inicialización de librerías (esto NO es parte del script demo_seg_orden1.py)
# Setup inline graphics: Esto lo hacemos para que el tamaño de la salida, 
# sea un poco más adecuada al tamaño del jupyter notebook documento

# algún bug cuando lo hice:
plt.figure(1)
plt.close(1)

import matplotlib as mpl

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 16

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})

#%%  Fin - Inicialización de librerías
```

##### Calculos y graficos 

```python
w0 = 10
qq = [0.1, 0.5, 2]

# Por cada valor de Q (qq) genero los graficos correspondientes

for ii in range(len(qq)):
      
    # Definicion de la funcion transferencia
    my_tf = TransferFunction( [w0**2], [1, w0/qq[ii], w0**2] )

    # Grafico de bode
    bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq[ii]) )
    
    # Grafico de polos y ceros
    pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq[ii])) #S plane pole/zero plot

    # Grafico temporal delay    
    GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq[ii]))

```

#### Punto 2 - Filtro PASA BANDA

A continuacion se muestra la ecuacion generica de un filtro pasabanda:

\begin{equation}
H(s) = \frac{s\frac{\omega_0}{Q}}{s^2+s\frac{\omega_0}{Q}+\omega_0^2}
\end{equation}

siendo $\omega_0$ la frecuencia de corte del filtro y $Q$ el factor de selectividad.

Ahora, para graficarlo le daremos los siguientes valores a la frecuencia y al factor de selectividad:

| $Q$ | $\omega_0$ | Tipo |
| :--- | :--- | :--- |
| $0.1$ | $10$ | Sobreamortiguado |
| $0.5$ | $10$ | Amortiguado crítico |
| $2$ | $10$| Subamortiguado |
| $5$ | $10$| Subamortiguado |
| $10$ | $10$| Subamortiguado |

A continuacion se procedera a realizar los graficos correspondientes mediante Python:

##### Calculos y graficos 

```python
w0 = 10
qq = [0.1, 0.5, 2, 5, 10]

# Por cada valor de Q (qq) genero los graficos correspondientes

for ii in range(len(qq)):
      
    # Definicion de la funcion transferencia
    my_tf = TransferFunction( [w0/qq[ii], 0], [1, w0/qq[ii], w0**2] )

    # Grafico de bode
    bodePlot(my_tf, fig_id=1, filter_description = 'Q={:3.3f}'.format(qq[ii]) )
    
    # Grafico de polos y ceros
    pzmap(my_tf, fig_id=2, filter_description = 'Q={:3.3f}'.format(qq[ii])) #S plane pole/zero plot

    # Grafico temporal delay    
    GroupDelay(my_tf, fig_id=3, filter_description = 'Q={:3.3f}'.format(qq[ii]))
```
